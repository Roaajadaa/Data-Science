```python
import pandas as pd 
import numpy as np
```

```python
md = pd.read_csv('missingdata.csv', )

md

```

```python
numnul = md.isnull().sum()
#number of missingdata in each coulm 
numnul
```

```python
# drop the coolumns that have 27000 or more than 27000 of missing data 
md = md.dropna( axis=1 , thresh = 27000)
md
```

```python
# drop the observation that have 100  or more than 100 of missing data 
md = md.dropna(axis=0 , thresh = 100)
```

```python
# reduse from 44596  to 43084 
md.isnull().sum().sum()
```

```python
 


headr = md.columns
#int64
#float64
listnum = []


#backward-filling for object detatype
liststr = []
for index in range(len(headr)):
    if (md[headr[index]].dtype == 'object'):
        liststr.append(headr[index])

for index3 in range(len(liststr)):
    md[liststr[index3]].fillna(method='bfill', inplace=True)
    
#each column have int or float datatype , fill missing data with mode

for index in range(len(headr)):
    if (md[headr[index]].dtype == 'int64' or md[headr[index]].dtype == 'float64'):
        listnum.append(headr[index])
        
for index2 in range(len(listnum)):
    md[listnum[index2]].fillna(md[listnum[index2]].mode()[0], inplace=True)

```

```python
#total missing data in data set after cleaning 

md.isnull().sum().sum()
```
